```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
from scipy import stats

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
data = pd.read_table('/home/kotarum/Ironhack/unit_7/lab-t-tests-p-values/files_for_lab/machine.txt', 
                    encoding='utf-16')
data
```

```python
data.columns = [column.lower().replace(" ", "_") for column in data.columns]
data
```

```python

t_stat, p_value = stats.ttest_ind(data['new_machine'], data['____old_machine'], equal_var=True)

one_sided_p_value = p_value / 2 if t_stat > 0 else 1 - p_value / 2

print("T-statistic:", t_stat)
print("One-sided P-value:", one_sided_p_value)

alpha = 0.05
if one_sided_p_value < alpha:
    print("We reject the null hypothesis. There is sufficient evidence that Old Machine is faster than New machine.")
else:
    print("We fail to reject the null hypothesis. There is not sufficient evidence that Old machine is faster than New machine.")
```

```python
data_2 = pd.read_table('/home/kotarum/Ironhack/unit_7/lab-t-tests-p-values/files_for_lab/student_gpa.txt', 
                    encoding='utf-8')
data_2.columns = [column.lower().replace(" ", "_") for column in data_2.columns]
data_2
```

```python
data_2['__juniors'] = data_2['__juniors'].fillna(data_2['__juniors'].mean())
data_2
```

```python
import pandas as pd
from scipy import stats


t_stat, p_value = stats.ttest_ind(data_2['sophomores'], data_2['__juniors'], equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis. There is sufficient evidence that the mean GPAs of sophomores and juniors differ.")
else:
    print("We fail to reject the null hypothesis. There is not sufficient evidence that the mean GPAs of sophomores and juniors differ.")
```
